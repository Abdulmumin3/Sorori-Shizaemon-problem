```python
grain = 1
grain_per_day = []
for day in range(100):
    if day == 0:
        grain_per_day.append(grain)
    else:
        grain_per_day.append((grain*2)**day)
print("The total rice grains after 100 days: ", sum(grain_per_day))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Grains per day")
plt.xlabel("Number of Days")
plt.ylabel("Number of Rice Grains")
plt.plot(grain_per_day)
plt.show()
```

```python
def compute_grains_per_days(days=100):
    grain = 1
    grain_per_day = []
    for day in range(days):
        if day == 0:
            grain_per_day.append(grain)
        else:
            grain_per_day.append((grain*2)**day)
    print(f"The total rice grains after {days} days:", sum(grain_per_day))

```

```python
compute_grains_per_days(809)
```

```python
grains_in_bowl = 20000
grains_per_day = grains_in_bowl*2

def compute_days(people, grains):
    total_grain_daily = people * grains_per_day
    days = grains/total_grain_daily
    print(f"{people} people can survive {days} days on {grains} grains")
compute_days(4, 100000000000)
```
